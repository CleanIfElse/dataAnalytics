```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
# creating a new column called person and grabbing the age, sex to generate person status
titanic_df['person'] = titanic_df[['Age', 'Sex']].apply(male_female_child, axis=1)
```

```python
titanic_df['Survived'] = titanic_df.Survived.map({0: 'no', 1: 'yes'})
```

```python
titanic_df.head()
```

```python
sb.factorplot('Survived', data=titanic_df, kind='count')
```

```python
# Great so we found out more people didn't survive than survive.
# Is it based on Pclass?
sb.factorplot('Pclass', 'Survived', data=titanic_df)
```

P class is making a difference if a person survives or not.

Lets examine this closely. Did the child, female or male survive

```python
sb.factorplot('Pclass', 'Survived', hue='person', data=titanic_df)
```

From what this looks like, female were able to survive the longest. Male and specially child had a difficult time surviving
